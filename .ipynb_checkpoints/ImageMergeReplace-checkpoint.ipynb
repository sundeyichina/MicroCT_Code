```python
#the image is the 3D tensor with: x, y, z 
def Merge(image, label1, label2):
    # this function merge the two labels for the image in both 2D and 3D, will return the image
    if label1 > label2:
        index = (image == label1)
        image[index] = label2
        return image#, label1
    else:
        index = (image == label2)
        image[index] = label1
        return image #, label2

def Replace(image, ref_image, ref_label):
    Max = image.max()
    index = (ref_image==ref_label)
    image[index] = Max+1
    return image

def Resort(image):
    #this funciton makes the image with continuous labels
    count = 0
    Max = image.max()
    for pointer in range(Max+1):
        index = (image == pointer)
        num = index.sum().item()
        if num == 0: #do not have such entry
            mask = (image == image.max())
            image[mask] = pointer
                
```
